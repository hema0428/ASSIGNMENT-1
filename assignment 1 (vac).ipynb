```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://amberstudent.com/blog/post/most-popular-sports-in-usa"
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
sportsname=soup.select("td:nth-child(1)")
```

```python
sportsname
```

```python
annualviews=soup.select("td:nth-child(2)")
```

```python
annualviews
```

```python
establishedyear=soup.select("td:nth-child(3)")
```

```python
establishedyear
```

```python
popularityrating=soup.select("td:nth-child(4)")
```

```python
popularityrating
```

```python
import pandas
```

```python
sportslist=pandas.DataFrame(sportsname,columns=["Sportsname"])
```

```python
sportslist
```

```python
sportslist["Establishedyear"]=establishedyear
```

```python
sportslist
```

```python
sportslist["Popularityrating"]=popularityrating
```

```python
sportslist
```

```python
sportslist["Annualviews"]=annualviews
```
